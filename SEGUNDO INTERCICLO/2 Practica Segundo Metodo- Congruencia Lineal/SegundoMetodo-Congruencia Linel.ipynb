```python
import collections
def semillaSiguiente(mult,xoAnterior,incremento,md):
    numero=mult*xoAnterior+incremento
    numXn=numero % md
    return numXn

def buscarSemilla(dato,lista):
    for x in range(0,len(lista)):
        #print(lista[x])
        if (lista[x]==dato):
               return True
```

```python
def genNumPseudoaleatorios(mult,incremento,md,iteraciones,semilla):
    a=mult
    b=incremento
    m=md
    iteracion=iteraciones
    Xo=semilla
    #datoes para la tabla
    Xn=0
    Un=0
    arregloSemillas=[]
    semillasRepetidas=[]

    if a>0 and b>0 and m>0:
        print('Proceso')
        for i in range(iteracion):
            if (i==0):
                Xn=Xo
                arregloSemillas.append(Xn)
            else:      
                Xn=semillaSiguiente(a,Xn,b,m)
                Xn=Xn
                Un=Xn/m
                print(Un)
                busquedaSemilla=buscarSemilla(Xn,arregloSemillas)
                if(busquedaSemilla==True):
                    semillasRepetidas.append(Xn)
                    print('Semilla se repite en la iteracion: ',i,'dato:',Xn)
                arregloSemillas.append(Xn)                                             
    else:
        print('El dato ingresao es incorrecto')                                 
    print('La Frecuencia es:')
    print('Sin incluir la aplicacion inicial:')
    counter=collections.Counter(semillasRepetidas)
    print(counter)
```

```python
#Datos de entrada
mult=5
incremento=1
md=9
iteraciones=70
semilla=1
print("Introduce el valor de la semilla:", semilla)
print("Introduce el valor del multiplicador:", mult )
print("Introduce el valor del modulo:", md)
print("Introduce el valor de las iteraciones: ", iteraciones)
genNumPseudoaleatorios(mult,incremento,md,iteraciones,semilla)
```
