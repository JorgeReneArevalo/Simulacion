```python
import wmi
w = wmi.WMI(namespace="root\wmi")
temp=w.MSAcpi_ThermalZoneTemperature()[0].CurrentTemperature
```

```python
OxC = 8
import cpuinfo
nproc = float(cpuinfo.get_cpu_info()['count'])
vel = float(cpuinfo.get_cpu_info()['hz_advertised_friendly'].split(" ")[0])
flops = nproc*vel*OxC
mflops=int(flops*1024)

import psutil
cpu_uso=psutil.cpu_percent()
ram=psutil.virtual_memory().total
ram_uso=psutil.virtual_memory().used
```

```python
valores =[temp, mflops, cpu_uso, ram, ram_uso,  175934, 31541, 7356, 742118, 11427]

def obtener_pos(digs):
    valor1 =0
    valor2 =0
    if digs%2 !=0:
        valor1 = int(digs/2)
        valor2 = int(digs/2)+1
    else:
        valor1 = int(digs/2)
        valor2 = int(digs/2)
    return valor1,valor2

def calcular_num(iters, val, digs):
    x0_semilla = int(val)
    aum = obtener_pos(digs)
    print(" Iteración ", " Xn ", " Xn*Xn", " Longitud "," Ui "," Rn ")
    for i in range(iters):
        xn2= x0_semilla**2
        lon = len(str(xn2))
        ui = str(xn2)[int(lon/2)-aum[0]:int(lon/2)+aum[1]]
        rn = int(ui)/10**digs
        print(i, " ", x0_semilla," ",xn2, " ", lon, " ",ui, " ", rn)
        x0_semilla=int(ui)
    print("   ")
    
iters = 16
digs = 4
for i in valores:
    print("i: ", i)
    calcular_num(iters, i, digs)
    print("  ")
```
